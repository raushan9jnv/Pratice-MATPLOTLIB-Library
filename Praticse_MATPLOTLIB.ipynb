```python
import  matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=(1,2,3,4,5)
y=(5,6,456,8,45)
plt.plot(x,y,color='Green',linewidth=4,linestyle='dotted',marker='D',markersize=10)
plt.xlabel('raushan')
plt.ylabel('kumar')
plt.title('My Name')
plt.show()
```

```python
x=(1,2,3,4,5)
y=(5,6,456,8,45)
plt.plot(x,y,'rD')
plt.xlabel('raushan')
plt.ylabel('kumar')
plt.title('My Name')
plt.show()
```

```python
plt.plot(x,y,color='Green',alpha=0.7)
```

```python
days=[1,2,3,4,5,6]
max_temp=[34,56,78,54,34,54]
min_temp=[34,23,12,34,56,77]
avg_temp=[12,34,56,78,55,43]
plt.plot(days,max_temp,label='Max_Temp',marker='+',linestyle=':',markersize=15)
plt.plot(days,min_temp,label='Min_Temp',marker='D')
plt.plot(days,avg_temp,label='Avg_Temp',marker='*')

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Temperatures in days')
#plt.legend(loc='upper right')
plt.legend(loc='best')
plt.grid()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,50,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,152],rwidth=0.95,color='green')
```

```python
blood_sugar_men = [113,85,50,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['men','women'])
plt.xlabel('Blood Sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')
plt.legend()
```
